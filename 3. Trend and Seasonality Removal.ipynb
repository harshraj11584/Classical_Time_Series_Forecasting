First we remove Trend from a plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

data=pd.read_csv('datasets/shampoo.csv')
```

```python
data.plot()
```

See that Data has an Increasing Trend

1. Detrending by Differencing was done in ARIMA (notebook 1)
2. Detrending by Model Fitting :

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
X = [i for i in range(0, data.shape[0])]
X = np.reshape(X, (len(X), 1))
model.fit(X,np.ravel(data['Sales']))
trend = model.predict(X)

data['Sales'].plot(label='Data')
plt.plot(trend, label='trend')

detrend = np.ravel(data['Sales'])-trend
plt.plot(detrend, label = 'detrended data')
plt.legend()
```

Will show Seasonality Removal using Another Dataset

```python
data = pd.read_csv('datasets/daily-min-temperatures.csv')
data.plot()
data.head()
```

See that this data has Seasonality. Converting dates to see the seasonality.

```python
data['Date'] = pd.to_datetime(data['Date'])
print(data.dtypes)
plt.plot(data['Temp'])
```

We will resample the dataset to a monhly average min temp to see the monthly trends

```python
series = pd.Series.from_csv('datasets/daily-min-temperatures.csv', header=0)

resample = series.resample('M')
monthly_mean = resample.mean()
print(monthly_mean.head(13))
monthly_mean.plot()
plt.show()
```

See that this clearly shows the seasonality of the dataset. We can subtract  this monthly mean to remove the seasonality of the dataset.

Next Method : 

Seasonal Adjustment using Modeling. 

We can first fit a model that clearly captures the Seasonality of the Dataset, then subtract the values to remove seasonality.

We try to fit polynomials of varying degree and see which captures the seasonality well.

```python
X = [i%365 for i in range(0, data.shape[0])]
y = np.ravel(data['Temp'])

fig=plt.figure(figsize=(16, 14), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(y[:730],label='data',color='y')
coeff, y_pred = [],[]
for d in range(2,5):
    coeff = np.polyfit(X,y, deg=d)
    y_pred = np.polyval(coeff, X)
    plt.plot(y_pred[:730],label='deg='+str(d))

plt.legend()
```

See that degree 4 and above sufficiently capture the Seasonality of the dataset

```python
deseasoned = y - y_pred
plt.plot(y[:730],label='data',color='y')
plt.plot(deseasoned[:730], label='deseasoned dataset')
plt.legend()
```

See that Seasonality of the DataSet has been Removed.

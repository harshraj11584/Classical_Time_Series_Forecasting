Moving Average Denoising is applied to STATIONARY time series to remove the fine-grained variation between time steps.

The hope is to remove noise and better expose the signal of the underlying causal processes. 

If we apply this method to non-stationary time series, we may lose vital information


1. Centered Moving Average

2. Trailing Moving Average

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('datasets/daily-total-female-births.csv')
data.plot()
n=data.shape[0]
print(data.shape)
```

As this data appears approx stationary, we apply Moving Average Denoising technique.

Iterating over different Window Widths to check best fits by plotting.

```python
plt.figure(figsize=(20,10))
plt.plot(data['Births'][:50],label='Data')
for w in range(2,5):
    obs = [sum([data['Births'][t-i] for i in range(w)])/w for t in range(w-1,n)]
    plt.plot(obs[:50],label='width='+str(w))
    
plt.legend()
```

```python
import pandas as pd
import numpy as np
from math import sqrt
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
%matplotlib inline

data= pd.read_csv('datasets/daily-min-temperatures.csv')
#Converting Date Column to pandas Date type
data['Date'] = pd.to_datetime(data['Date'])

n=data.shape[0]
print(data.head())
print(data.shape)
print(data.count(axis=0))
print(type(data['Date'][0]))
print(type(data['Temp'][0]))
```

Establishing Baseline RMSE

```python
def model_persistence(x):
    return (x)

baseline_predictions = np.array([model_persistence(data["Temp"][i]) for i in range(n-1)])
actual_baseline = np.array([data["Temp"][i] for i in range(1,n)])

rmse_baseline = sqrt(mean_squared_error(actual_baseline, baseline_predictions))
print("Baseline Pred RMSE = ", rmse_baseline)
```

See Plot of the DataSet

```python
data['Temp'].plot()
```

Trend → a general systematic linear or (most often) nonlinear component that changes over time and does not repeat

Seasonality → a general systematic linear or (most often) nonlinear component that changes over time and does repeat

Noise → a non-systematic component that is nor Trend/Seasonality within the data

1. Date-Time Features

Integer Month and Day may help extract seasonality Information

Changing dataset [Date,Temp] -> [Year, Month, Day, Temp] 

```python
data_ymdt = pd.DataFrame()
data_ymdt['Year'] = [data['Date'][i].year for i in range(n)]
data_ymdt['Month'] = [data['Date'][i].month for i in range(n)]
data_ymdt['Day'] = [data['Date'][i].day for i in range(n)]

data_ymdt['Temp'] = data['Temp']

print(data_ymdt.head())
```

2. Lag Features

Changing the Dataset to include Lag Features to predict Current Temperature

```python
#AutoCorrelation Plot
import statsmodels.graphics.tsaplots
statsmodels.graphics.tsaplots.plot_acf(data['Temp'])
```

Checking how many lag points to include from current year

```python
from statsmodels.tsa.ar_model import AR
AR_model = AR(data["Temp"])
print("Check Max Lag using Akaike Information Criterion")
print("\nFitting")
modelAR_fit = AR_model.fit(maxlag=366, method='cmle', ic='aic', trend='c', maxiter=1000)
print("Lag = ", modelAR_fit.k_ar)
```

21 Lag Points give Sufficient Results

Including 21 Lag Points in Dataset

```python
temps = pd.DataFrame(data["Temp"])
data_lag_inc = pd.DataFrame()
for i in range(1,22):
    data_lag_inc = pd.concat([data_lag_inc, temps.shift(i)],axis=1)
data_lag_inc.columns = ["temp"+str(i) for i in range(1,22)]
data_lag_inc = data_lag_inc.dropna()
print(data_lag_inc.shape)
print(data_lag_inc.head())
```

Including Date Time Features and Lag Features Together in Dataset

```python
data1 = pd.concat([data_ymdt, data_lag_inc], axis=1)
data1 = data1.dropna()
print(data1.shape)
print(data1.head(2))
```

Creating Test/Train DataSets

```python
data_train = data1.iloc[:3000,:]
data_test = data1.iloc[3000:,:]

x_train = data_train.drop('Temp', 1)
print(x_train.head())
y_train = pd.DataFrame(data_train['Temp'])

x_test = data_test.drop('Temp', 1)
print(x_test.head())
y_test = pd.DataFrame(data_test['Temp'])

print("x_test: ", x_test.shape)
print("x_train: ", x_train.shape)
print("y_test: ", y_test.shape)
print("y_train: ", y_train.shape)
```

Trying Some Generalized Linear Models

1. Ordinary Least Squares

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(x_train, y_train)
coeff_det = reg.score(x_train, y_train)
pred1 = reg.predict(x_test)
rmse1 = sqrt(mean_squared_error(y_test, pred1))

pred = reg.predict(x_train)
print("train rmse= ",sqrt(mean_squared_error(y_train, pred)))

print("R^2=", coeff_det)
print("rmse=", rmse1)
```

2. Ridge Regression

```python
from sklearn import linear_model
reg = linear_model.Ridge(alpha=0.1, max_iter=100000)
reg.fit(x_train, y_train)
coeff_det = reg.score(x_train, y_train)

pred = reg.predict(x_train)
print("train rmse= ",sqrt(mean_squared_error(y_train, pred)))

pred2 = reg.predict(x_test)
rmse2 = sqrt(mean_squared_error(y_test, pred2))

print("R^2=", coeff_det)
print("rmse=", rmse2)
```

3. LASSO

```python
from sklearn import linear_model
reg = linear_model.Lasso(alpha=0.01, max_iter=100000)
reg.fit(x_train, y_train)
coeff_det = reg.score(x_train, y_train)

pred = reg.predict(x_train)
print("train rmse= ",sqrt(mean_squared_error(y_train, pred)))

pred3 = reg.predict(x_test)
rmse3 = sqrt(mean_squared_error(y_test, pred3))

print("R^2=", coeff_det)
print("rmse=", rmse3)
```

Bayesian Ridge Regression

```python
from sklearn import linear_model
reg = linear_model.BayesianRidge(n_iter=10000)
reg.fit(x_train, np.ravel(y_train))
coeff_det = reg.score(x_train, np.ravel(y_train))

pred = reg.predict(x_train)
print("train rmse= ",sqrt(mean_squared_error(y_train, pred)))

pred4 = reg.predict(x_test)
rmse4 = sqrt(mean_squared_error(y_test, pred4))

print("R^2=", coeff_det)
print("rmse=", rmse4)
```

Plotting First 50 Predictions

```python
plt.plot(np.linspace(0,1,50),y_test[:50],label="y_test")
plt.plot(np.linspace(0,1,50),pred1[:50],label="pred1")
plt.plot(np.linspace(0,1,50),pred2[:50],label="pred2")
plt.plot(np.linspace(0,1,50),pred3[:50],label="pred3")
plt.plot(np.linspace(0,1,50),pred4[:50],label="pred4")

plt.legend()
```

See that all these Simple Linear Methods Give Same Predictions

Try to Fit a for Degree 2 Now

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
x_train_transf = poly.fit_transform(x_train)
x_test_transf = poly.fit_transform(x_test)
```

Trying Bayesian Ridge Regression Here

```python
from sklearn import linear_model
reg = linear_model.BayesianRidge(n_iter=10000)
reg.fit(x_train_transf, np.ravel(y_train))
coeff_det = reg.score(x_train_transf, np.ravel(y_train))

pred = reg.predict(x_train_transf)
print("train rmse= ",sqrt(mean_squared_error(y_train, pred)))

pred5 = reg.predict(x_test_transf)
rmse5 = sqrt(mean_squared_error(y_test, pred5))

print("R^2=", coeff_det)
print("rmse=", rmse5)
```

```python
plt.plot(np.linspace(0,1,50),y_test[:50],label="y_test")
plt.plot(np.linspace(0,1,50),pred4[:50],label="pred4")
plt.plot(np.linspace(0,1,50),pred5[:50],label="pred5")
plt.legend()
```

Trying with Degree 3

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge

model = Pipeline([('poly', PolynomialFeatures(degree=3)),
                  ('linear', Ridge(alpha=1000.0,fit_intercept=True))])
model.fit(x_train,y_train)
coeff_det = model.score(x_train, np.ravel(y_train))

pred = model.predict(x_train)
print("train rmse= ",sqrt(mean_squared_error(y_train, pred)))

pred6 = model.predict(x_test)
rmse6 = sqrt(mean_squared_error(y_test, pred6))

print("R^2=", coeff_det)
print("rmse=", rmse6)
```

```python
plt.plot(np.linspace(0,1,50),y_test[:50],label="y_test")
plt.plot(np.linspace(0,1,50),pred5[:50],label="pred5")
plt.plot(np.linspace(0,1,50),pred6[:50],label="pred6")
plt.legend()
```

Trying Simple MLP

```python
from sklearn.neural_network import MLPRegressor as MLP
model = MLP(
    hidden_layer_sizes=(200,100, ), activation='relu', 
    solver='adam', alpha=0.0001, batch_size='auto', 
    learning_rate='constant', learning_rate_init=0.001, 
    max_iter=100000, shuffle=False, tol=0.000001, momentum=0.9, 
    nesterovs_momentum=True, early_stopping=False, 
    validation_fraction=0.1, beta_1=0.9, 
    beta_2=0.999, epsilon=1e-08, n_iter_no_change=10)
model.fit(x_train, np.ravel(y_train))
pred = model.predict(x_train)
print("train rmse= ",sqrt(mean_squared_error(y_train, pred)))

pred7 = model.predict(x_test)
rmse7 = sqrt(mean_squared_error(y_test, pred7))

print("rmse=", rmse7)
```

```python
plt.plot(np.linspace(0,1,50),y_test[:50],label="y_test")
plt.plot(np.linspace(0,1,50),pred5[:50],label="pred5")
plt.plot(np.linspace(0,1,50),pred6[:50],label="pred6")
plt.plot(np.linspace(0,1,50),pred7[:50],label="pred7")
plt.legend()
```
